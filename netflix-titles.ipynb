# Data Cleaning and EDA

**Printing the first rows to get an overview on the data**

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
df.head()
```

**Printing all columns**

```python
df.columns
```

**An Overview**

```python
df.shape  # Check number of rows and columns
df.info()  # Check data types and null values
df.describe(include='all')  # Summary stats (use include='all' to see object types)
```

**Check how many missing values each column**

```python
df.isnull().sum()
```

**Droping rows with missing values**

```python
df.dropna(inplace=True)
```

**Removing Duplicates**

```python
df.drop_duplicates(inplace=True)
```

**Checking**

```python
df.shape  # Check number of rows and columns
df.info()  # Check data types and null values
df.describe(include='all')  # Summary stats (use include='all' to see object types)
```

**Choosing relevant columns**

```python
columns_to_use = ['type',  'country', 'date_added',
                  'release_year', 'rating', 'duration', 'listed_in']
df = df[columns_to_use]
```

```python
df['country'].unique
```

```python
# Split entries with multiple countries and explode into separate rows
df['country'] = df['country'].apply(lambda x: x.split(', '))
df_exploded = df.explode('country')
country_counts = df_exploded['country'].value_counts()
```

**Plot a Pie Chart of the Top 10 Countries**

```python
import matplotlib.pyplot as plt

top_countries = country_counts.head(10)

plt.figure(figsize=(8, 8))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Countries Producing Netflix Content')
plt.axis('equal')  # Ensures the pie is circular
plt.show()
```

```python
type_country_counts = df_exploded.groupby(['country', 'type']).size().unstack(fill_value=0)
```

**Plot a Horizontal Bar Chart (Top 10 Countries)**

```python
# Get top 10 countries by total titles
top_countries = type_country_counts.sum(axis=1).sort_values(ascending=False).head(10)
top_data = type_country_counts.loc[top_countries.index]

# Plot
top_data.plot(kind='barh', stacked=True, figsize=(10, 6), colormap='Set2')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.title('Movies vs TV Shows by Country (Top 10)')
plt.legend(title='Type')
plt.gca().invert_yaxis()  # Most common country on top
plt.tight_layout()
plt.show()
```

**Top genres (from listed_in)**

```python
# Split multiple genres
df['genres'] = df['listed_in'].apply(lambda x: x.split(', '))
df_genres = df.explode('genres')
df_genres['genres'].value_counts().head(10).plot(kind='bar', title='Top 10 Genres')
```
